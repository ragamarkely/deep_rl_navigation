Environments contain **_brains_** which are responsible for deciding the actions of their associated agents. Here we check for the first brain available, and set it as the default brain we will be controlling from Python.

````python
# Get the default brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
````

### 2. Examine the State and Action Spaces

The simulation contains a single agent that navigates a large environment.  At each time step, it has four actions at its disposal:
- `0` - walk forward 
- `1` - walk backward
- `2` - turn left
- `3` - turn right

The state space has `37` dimensions and contains the agent's velocity, along with ray-based perception of objects around agent's forward direction.  A reward of `+1` is provided for collecting a yellow banana, and a reward of `-1` is provided for collecting a blue banana. 

Run the code cell below to print some information about the environment.

````python
# Get the environment info
env_info = env.reset(train_mode=True)[brain_name]
action_size = brain.vector_action_space_size
state = env_info.vector_observations[0]
state_size = len(state)
````

### 3. Train a Deep Reinforcement Learning Agent

When training the environment, set `train_mode=True`, so that the line for resetting the environment looks like the following:
```python
env_info = env.reset(train_mode=True)[brain_name]
```

````python
WINDOW_SIZE = 100  # number of episodes to calculate the latest average score.

def train(
    agent: Agent,
    env: UnityEnvironment,
    n_episodes: int = 2000, 
    max_t: int = 1000, 
    eps_start: float = 1.0, 
    eps_end: float = 0.01, 
    eps_decay: float = 0.99
) -> Tuple[List[float], List[float]]:
    """
    Train RL agent.
    
    Params
    ======
    agent: RL Agent
    env: environment for interacting with and training the agent
    n_episodes: max number of training episodes
    max_t: max number of timesteps per episodes
    eps_start: starting value of epsilon for epsilon-greedy action selection
    eps_end: minimum value of epsilon
    eps_decay: decay factor per episode for decreasing epsilon
    
    Returns
    =======
    scores: list of scores from each episode
    scores_window: last WINDOW_SIZE scores
    """
    solved = False
    scores = []                        
    scores_window = deque(maxlen=WINDOW_SIZE)  
    eps = eps_start
    brain_name = env.brain_names[0]
    
    for i_episode in range(1, n_episodes + 1):
        state = env.reset(train_mode=True)[brain_name].vector_observations[0]
        score = 0
        for _ in range(max_t):
            action = agent.act(state, eps)                 # select an action
            env_info = env.step(action)[brain_name]        # send the action to the environment
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]                  # see if episode has finished
            
            agent.step(state, action, reward, next_state, done)
            score += reward                                # update the score
            state = next_state                             # roll over the state to next time step
            if done:
                break
                
        scores.append(score)
        scores_window.append(score)
        avg_score = np.mean(scores_window)
        eps = max(eps_end, eps_decay * eps)
        if i_episode % WINDOW_SIZE == 0:
            print(f"Episode {i_episode}. Average Score: {avg_score}.")
        if not solved and avg_score >= 13.0:
            print(f"Environment is solved in {i_episode} episodes! Average score: {avg_score}")
            torch.save(agent.qnetwork.state_dict(), "checkpoint.pth")
            solved = True
            
    if avg_score >= 13.0:
        print(f"Saving Q network parameters at the end of training.")
        torch.save(agent.qnetwork.state_dict(), "checkpoint.pth")
            
    return scores, scores_window
````

````python
# Get the agent
# A Double DQN agent (imported from agent.py) with experience replay and fixed Q target 
# is used for this project.
# Each of the Q-Network consists of 3 fully connected layers.
# The first two hidden layers has 64 neurons, and 
# the number of neurons in the output layer is equal to the action size.
# The other hyperparameters are defined in agent.py
agent = Agent(state_size=state_size, action_size=action_size, seed=888)

# Train the agent
scores, scores_window = train(agent=agent, env=env)
````

````python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(scores)
ax.set_xlabel("Episode")
ax.set_ylabel("Reward");
````

````python
# Close the environment
env.close()
````

## Future improvement

Several ideas for improving the agent performance:
1. Increasing the number of hidden layers and neurons. Several values were evaluated for the number of neurons in this project, and 64 were sufficient to solve the environment within a reasonable number of episodes. Higher number of neurons require more episodes to solve the environment. However, if our goal is to achieve the highest score, we could compromise on the number of episodes required for training.
2. Although the agent successfully solves the environment, there are still significant amount of fluctuation in the total rewards in the last 100 episodes. It will be interesting to compare the performance with a double DQN that learns directly from pixel, and investigate whether the significant fluctuation is still present. Running the training for significantly higher number of episodes could also be explored for reducing the noise.
3. Try other value-based algorithms, such as dueling DQN, prioritized experience replay, multistep bootstrap targets, distributional DQN, noisy DQN, and rainbow.
